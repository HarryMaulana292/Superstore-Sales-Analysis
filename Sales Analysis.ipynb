```python
df_YProfit = pd.concat([df["Order Date"].dt.year, df["Order Date"].dt.month, df["Profit"]], axis=1)
df_YProfit.columns = ["Year", "Month", "Profit"]
display(df_YProfit, df_YProfit.groupby("Year")[["Profit"]].sum())
```

```python
df_dummy = df_YProfit.groupby("Year")["Profit"].sum().reset_index()
def growth_profit(df):
    growth = [
        None if i == 0 else round(((df.loc[i, "Profit"] - df.loc[i-1, "Profit"]) / df.loc[i-1, "Profit"]) * 100, 2)
        for i in range(len(df))
    ]
    
    df["Growth (%)"] = growth
    return df
df_dummy["Profit"].values
```

```python
df_dummy = df_YProfit.groupby("Year")["Profit"].sum().reset_index()
# df_dummy["Profit"][1]-df_dummy["Profit"][0]
def inc(df):
    x = [0 if i == 0 else float(df[i]-df[i-1]) for i in df.index]
    return [0 if i == 0 else round((j/df[i-1])*100) for i,j in enumerate(x)]
inc(df_dummy["Profit"])
```

Berdasarkan grafik diatas, kita dapat mengetahui bahwa profit Superstore selalu mengalami peningkatan di tiap tahunnya. Di tahun 2015, keuntungan meningkat 24% dari profit tahun sebelumnya, dan di tahun berikutnya (2016) juga mengalami peningkatan 33% dari profit 2015. Namun di tahun 2017, tercatat hanya 14% peningkatan profit dari tahun 2016.

Jika di breakdown by year, dapat dilihat bahwa profit selalu mengalami fluktuasi tiap tahunnya. Tercatat di Juli 2014 dan Januari 2015, Superstore sempat mengalami kerugian. Disisi lain, hanya profit di bulan Februari dan bulan April yang keuntungannya sangat sulit menembus angka 5000. Sementara itu, performa penjualan setiap bulan September selalu naik, dan sebaliknya di bulan Januari penjualan selalu menurun.

Analysis:
* Berdasarkan polanya, penurunan yang terjadi pada Juli 2014 kemungkinan besar dipengaruhi oleh pola musiman di industri ritel. Berdasarkan data dari [U.S. Census Bureau](https://www.bea.gov/data/consumer-spending/main), bulan Juli umumnya berada di luar periode peak season, sehingga aktivitas belanja cenderung lebih stabil atau bahkan menurun dibandingkan bulan-bulan dengan promosi besar. Selain itu, menurut analisis industri ritel yang dirilis oleh [National Retail Federation](https://nrf.com/resources/holiday-and-seasonal-trends), pada periode musim panas, sebagian konsumen di US cenderung mengalihkan pengeluaran mereka ke aktivitas rekreasi, perjalanan, atau liburan dibandingkan pembelian produk ritel tertentu. Kondisi ini dapat menyebabkan perlambatan sementara dalam performa penjualan. Hal ini menjelaskan mengapa pada Juli 2014 Superstore mengalami kerugian.

* Penurunan performa pada Januari 2015 kemungkinan besar dipengaruhi oleh faktor musiman, khususnya fenomena post-holiday effect, di mana tingkat consumer spending fatigue masyarakat di US cenderung menurun setelah puncak belanja di Black Friday dan Christmas. Menurut [National Retail Federation](https://nrf.com/), tren ini merupakan pola tahunan di mana konsumen umumnya mengurangi pengeluaran untuk melakukan financial recovery serta memenuhi kewajiban pembayaran setelah liburan. Selain itu, di awal 2015 beberapa wilayah AS mengalami winter storm sehingga mempengaruhi aktivitas belanja, terutama pada sektor ritel offline. Hal ini menjelaskan kenapa pada Januari 2015 Superstore mengalami kerugian sekaligus menjawab mengapa periode Desember-Januari, performa penjualan selalu menurun.

* Terjadi peningkatan penjualan secara konsisten pada bulan September di setiap tahunnya, yang mengindikasikan adanya pola musiman (seasonality). Hal ini kemungkinan dipengaruhi oleh periode back-to-school di Amerika Serikat, di mana permintaan terhadap produk tertentu seperti perlengkapan kantor dan teknologi meningkat. Selain itu, September juga menandai berakhirnya liburan musim panas, sehingga aktivitas ekonomi dan konsumsi masyarakat kembali meningkat. 

* Profit pada bulan Februari dan April secara konsisten menunjukkan performa yang relatif rendah dan sulit melampaui angka 5000. Hal ini mengindikasikan adanya pola musiman yang mempengaruhi profitabilitas pada kedua bulan tersebut. Penurunan performa penjualan setiap bulan Februari pada dasarnya masih terdampak oleh post-holiday effect. Sementara itu, pada bulan April, faktor tax season di US (Deadline pajak biasanya pertengahan April) berpotensi menekan daya beli konsumen karena adanya kewajiban pembayaran pajak.

Dengan ini, terjawablah beberapa pertanyaan berikut.
* Bagaimana perkembangan profit Superstore dari tahun ke tahun?
    * Meskipun performa penjualan di tiap bulan selalu mengalami fluktuasi, profit Superstore selalu meningkat tiap tahunnya.
* Pada kondisi seperti apa profit selalu mengalami kenaikan?
    * Event puncak (seperti Black Friday dan Christmas) dan periode back-to-school (khususnya di bulan September) yang menandai berakhirnya liburan musim panas berpeluang meningkatkan penjualan.
* Pada kondisi seperti apa profit selalu mengalami penurunan?
    * Penurunan performa penjualan dapat disebabkan beberapa faktor musiman seperti post-holiday effect, tax season, kondisi cuaca (winter storm), dan peralihan pengeluaran ke aktivitas rekreasi di liburan musim panas.

Rekomendasi:
* Optimasi Low Season (Januari, Februari, April, Juli) dengan mengunakan promosi terarah (bundling / diskon terbatas) serta efisiensi stok untuk menjaga demand tanpa menekan margin.
* Maksimalkan High Season (September) dengan meningkatkan ketersediaan produk (terutama Technology & Office Supplies) dan optimalkan pricing tanpa ketergantungan pada diskon besar.
* Terapkan Strategy Berbasis Seasonality dengan menyesuaikan pricing, inventory, dan campaign marketing dengan pola musiman untuk menjaga stabilitas penjualan dan memaksimalkan profit sepanjang tahun.
